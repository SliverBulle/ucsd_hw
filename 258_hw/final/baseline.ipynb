```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shuyangli94/food-com-recipes-and-user-interactions")

print("Path to dataset files:", path)
```

```python
import pandas as pd

recipes = pd.read_csv(f"{path}/PP_recipes.csv")
recipes.head()
```

```python
users = pd.read_csv(f"{path}/PP_users.csv")
```

```python
users.head()
```

```python
raw_interactions = pd.read_csv(f"{path}/RAW_interactions.csv")
raw_interactions.head()
```

```python
raw_recipes = pd.read_csv(f"{path}/RAW_recipes.csv")
raw_recipes.head()
```

```python
interactions_test = pd.read_csv(f"{path}/interactions_test.csv")
interactions_test.head()
```

```python
interactions_train = pd.read_csv(f"{path}/interactions_train.csv")
interactions_train.head()
```

```python
interactions_validation = pd.read_csv(f"{path}/interactions_validation.csv")
interactions_validation.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
```

```python
# 1. 特征工程
def create_features(interactions_df, recipes_df, users_df):
    # 计算用户的平均评分
    user_avg_rating = interactions_df.groupby('u')['rating'].mean().reset_index()
    user_avg_rating.columns = ['u', 'user_avg_rating']
    
    # 计算食谱的平均评分
    recipe_avg_rating = interactions_df.groupby('i')['rating'].agg(['mean', 'count']).reset_index()
    recipe_avg_rating.columns = ['i', 'recipe_avg_rating', 'recipe_rating_count']
    
    # 合并特征
    df = interactions_df.merge(user_avg_rating, on='u', how='left')
    df = df.merge(recipe_avg_rating, on='i', how='left')
    
    return df

# 2. 准备训练数据
def prepare_data(df):
    features = ['user_avg_rating', 'recipe_avg_rating', 'recipe_rating_count']
    X = df[features]
    y = df['rating']
    return X, y

# 3. 训练模型
def train_model(X_train, y_train):
    model = LinearRegression()
    model.fit(X_train, y_train)
    return model

# 4. 评估函数
def evaluate_model(model, X, y):
    predictions = model.predict(X)
    # 将预测值限制在1-5之间
    predictions = np.clip(predictions, 1, 5)
    
    rmse = np.sqrt(mean_squared_error(y, predictions))
    mae = mean_absolute_error(y, predictions)
    return rmse, mae

# 主流程
# 1. 准备数据
train_df = create_features(interactions_train, raw_recipes, users)
val_df = create_features(interactions_validation, raw_recipes, users)
test_df = create_features(interactions_test, raw_recipes, users)

# 2. 准备特征
X_train, y_train = prepare_data(train_df)
X_val, y_val = prepare_data(val_df)
X_test, y_test = prepare_data(test_df)

# 3. 训练模型
model = train_model(X_train, y_train)

# 4. 评估模型
train_rmse, train_mae = evaluate_model(model, X_train, y_train)
val_rmse, val_mae = evaluate_model(model, X_val, y_val)
test_rmse, test_mae = evaluate_model(model, X_test, y_test)

# 打印结果
print("训练集结果：")
print(f"RMSE: {train_rmse:.4f}")
print(f"MAE: {train_mae:.4f}")
print("\n验证集结果：")
print(f"RMSE: {val_rmse:.4f}")
print(f"MAE: {val_mae:.4f}")
print("\n测试集结果：")
print(f"RMSE: {test_rmse:.4f}")
print(f"MAE: {test_mae:.4f}")

# 打印特征重要性
features = ['用户平均评分', '食谱平均评分', '食谱评分数量']
for feature, coef in zip(features, model.coef_):
    print(f"{feature}的系数: {coef:.4f}")
print(f"截距: {model.intercept_:.4f}")
```
